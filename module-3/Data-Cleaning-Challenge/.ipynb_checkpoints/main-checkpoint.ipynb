```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris-data.csv')
df.head()
```

```python
df.info()
```

There are 5 values missed in 'petal_width_cm', park for now, will deal with it later, after high level review of the whole dataset.

```python
df.describe()
```

All numeric values look reasonable, although minimum values of 'sepal_length' might be investigated further as it varies to much from mean value.

```python
df['class'].unique()
```

Class names should be aligned.

```python
old_names = ['Iris-setossa', 'versicolor']
new_names = ['Iris-setosa', 'Iris-versicolor']
df['class'].replace(old_names, new_names, inplace=True)
df['class'].unique()
```

**Let's deal with missing values now.**

```python
df.loc[df['petal_width_cm'].isnull()]
```

```python
len(df.loc[df['petal_width_cm'].isnull()])/len(df)*100
```

3% is not very much so we could drop these rows but let's check first if there is a way to add missing values.

```python
df.loc[df['class'] == 'Iris-setosa']
```

```python
sns.heatmap(df.loc[df['class'] == 'Iris-setosa'].corr(), annot=True)
```

There is no correlation between petal_width and other measurements so let's use an average to fill in missing data.

```python
mean_ = df.loc[df['class'] == 'Iris-setosa'].groupby('class').mean()['petal_width_cm']
mean_
```

```python
df.fillna(mean_[0], inplace=True)
df.info()
```

```python
df['class'].hist()
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(data=df)
plt.show()
```

Outliers of sepal_length looks like an error rather than a relevant data.

```python
df.sort_values('sepal_length_cm')[:10]
```

There are five values that don't look reasonable. I will replace them with average.

```python
mean_sl = df.loc[df['class'] == 'Iris-versicolor'].groupby('class').mean()['sepal_length_cm']
mean_sl
```

```python
df['sepal_length_cm'] = np.where(df['sepal_length_cm'] < 1, mean_sl, df['sepal_length_cm'])
df['sepal_length_cm'].min()
```

**Overall review once again after changes made**

```python
df.describe()
```

```python
sns.boxplot(data=df)
plt.show()
```
