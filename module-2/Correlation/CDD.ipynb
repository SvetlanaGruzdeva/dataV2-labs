```python
import requests
from io import StringIO
import pandas as pd
import seaborn as sns
```

```python
orig_url = 'https://drive.google.com/file/d/10bzUmn-E5LKxi2APB-mPbm-X330X8H2-/view?usp=sharing'

file_id = orig_url.split('/')[-2]
dwn_url = 'https://drive.google.com/uc?export=download&id=' + file_id
url = requests.get(dwn_url).text
csv_raw = StringIO(url)
cdd = pd.read_csv(csv_raw, sep='\t', header=0)
cdd.head(10)
```

```python
print(cdd.shape)

cdd.info()
# All numeric fields contains some of NAN values. To have more readable data I will replace it with zero.
```

```python
cdd = cdd.fillna(0)
cdd.head(10)
```

```python
# Correlation matrix
cdd.corr()
```

```python
# Heat map
cdd.corr().style.background_gradient(cmap='coolwarm')
```

**3.** Find 3 pairs of columns, one pair that is highly positively correlated, one for which the correlation is almost zero and one pair that is negatively correlated. In each case write down an explanation why we see this correlation.

```python
# Positively correlated pair: fatalities & injured/infected.
print(cdd['FATALITIES'].corr(cdd['INJURED / INFECTED']))
'''
These two data sets have the same reasons of cause but they don't relate to each other directly. The greater the disaster is
the more casualties it will cause - both in deaths and injuries.
'''
```

```python
# Non-correlated pair: magnitude & insurange payments
print(cdd['MAGNITUDE'].corr(cdd['INSURANCE PAYMENTS']))
'''
I would expect strong positive correlation between this two parameters because I would expect this relation: the higher the
magnitude => the stronger is disaster => the greater are consecuentes (both for prorerty and people) => the higher insurance
payments.
One reason for very little correlation might be that real disasters are actually not covered by insurance. I'm sure there are
some special types of insurance even for disasters but if this type is not included into our population then there will be
almost no correlation between magnitude and insurance payments.
Another reason we can assume if we look closer into data set provided. As per short analysis below for most of events magnitude
is not filled in because it's not relevant for these types of disaster.
'''
```

```python
cdd.groupby('EVENT CATEGORY')['EVENT CATEGORY'].count()
```

```python
cdd.groupby('MAGNITUDE')['EVENT CATEGORY'].count()
```

```python
cdd.groupby('EVENT SUBGROUP')['EVENT CATEGORY'].count()
```

```python
pd.DataFrame(cdd.loc[cdd['MAGNITUDE'] > 0, 'EVENT SUBGROUP']).groupby('EVENT SUBGROUP')['EVENT SUBGROUP'].count()
```

```python
# Negatively correlated pair: magnitude & provincial department payment (Note: in provided population all negative correlations
# found are very close to zero (>-0.01))
```

```python
print(cdd['MAGNITUDE'].corr(cdd['PROVINCIAL DEPARTMENT PAYMENTS']))
'''
I picked this value up because it's the lowest of negative correlations identified but it's still very close to zero.
Looking into short analysis below I can't really see much of relation between this two parameters. For example, out of 86
registered provincial payments only two relates to Geological type of disaster while almost 100% (11 out of 12) events where
magnitude has been registered are Geological.
'''
```

```python
cdd.groupby('PROVINCIAL DEPARTMENT PAYMENTS')['EVENT CATEGORY'].count()
```

```python
pd.DataFrame(cdd.loc[cdd['PROVINCIAL DEPARTMENT PAYMENTS'] > 0, 'EVENT SUBGROUP']).groupby('EVENT SUBGROUP')['EVENT SUBGROUP'].count()
```
