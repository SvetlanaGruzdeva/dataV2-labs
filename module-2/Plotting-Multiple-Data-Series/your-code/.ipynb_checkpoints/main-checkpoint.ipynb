# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
df = data[['ItemType', 'RetailSales', 'RetailTransfers', 'WarehouseSales']].groupby(data['ItemType']).sum().reset_index()

df.plot.bar(x='ItemType', figsize=(6,6))
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
# How I understand requirements - we should show each type of sales separately in a plot but evaluate suppliers by sum of them.
df = data.groupby(data['Supplier']).sum().reset_index()
df['TotalSales'] = df['RetailSales'] + df['RetailTransfers'] + df['WarehouseSales']
df = df.sort_values('TotalSales', ascending=False).head(10)
df.drop(['Year', 'Month'], axis=1, inplace=True)
df[['Supplier', 'RetailSales', 'RetailTransfers', 'WarehouseSales']].plot.barh('Supplier')
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
data['YearMonth'] = data['Year'].astype(str) + data['Month'].astype(str)
data['YearMonth'] = data['YearMonth'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m')).dt.to_period('M')
```

```python
df = data[['YearMonth', 'RetailSales', 'RetailTransfers', 'WarehouseSales']].groupby(data['YearMonth']).mean().reset_index()
df.plot(x='YearMonth', figsize=(8,8))
```

## 4. Plot the same information as above but as a bar chart.

```python
df.plot.bar(x='YearMonth', figsize=(8,8))
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
df = data[['YearMonth', 'ItemType', 'RetailSales']].pivot_table(index=['ItemType', 'YearMonth'], aggfunc='sum').reset_index()
```

```python
figure, ax = plt.subplots()
color = ['b', 'k', 'c', 'g', 'm', 'r', 'y', 'DarkOrange']
n = 0
for i in df['ItemType'].unique():
    df.loc[df['ItemType'] == i].plot(x='YearMonth', figsize=(8,8), color=color[n], ax=ax)
    plt.legend(i)
    n += 1

# There should be more efficient way, also legend works really wierd.
# UPD. Found that way right below, will keep current solution for history.
```

```python
df = df.pivot(index='YearMonth', columns='ItemType', values='RetailSales')
```

```python
df.plot()
```

## 6. Plot the same information as above but as a bar chart.

```python
df.plot.bar(width=1, figsize=(8,8))
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.lmplot(x='RetailSales', y='RetailTransfers', data=data, hue='ItemType')
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
sns.pairplot(data=data, hue='ItemType')
```
