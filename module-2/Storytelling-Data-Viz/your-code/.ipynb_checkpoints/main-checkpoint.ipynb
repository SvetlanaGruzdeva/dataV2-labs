## Bonus Challenge 1 - Adjust Data Distribution

If you used the correct method in the previous step, you should have found the data distribution is skewed to the left. In order to improve your data visualization in the next steps, you can opt to adjust the `SalePrice` column by applying a mathematical function to the values. The goal is to produce a bell-shape normal distribution after applying the mathematical function to the sale price.

*This technique is optional in data visualization but you'll find it useful in your future machine learning analysis.*

#### In the cell below, adjust the `SalePrice` column so that the data are normally distributed.

Try applying various mathematical functions such as square root, power, and log to the `SalePrice` column. Visualize the distribution of the adjusted data until you find a function that makes the data normally distributed. **Create a new column called `SalePriceAdjusted` to store the adjusted sale price.**

[This reference](https://trainingdatascience.com/workshops/histograms-and-skewed-data/) shows you examples on how to adjust skewed data.

```python
# your code here

df['SalePriceAdjusted'] = df['SalePrice'].apply(lambda x: math.log(x))

df['SalePriceAdjusted'].hist(bins=100, figsize=(6,6))
```

```python
# 3 cells below are about trying to draw normal distrinution line above my histogram.

df_pow = df['SalePrice'].apply(lambda x: math.log(x))
df_pow.std()
```

```python
import scipy.stats as stats 

param = stats.norm.fit(df_pow)   # Fit a normal distribution to the data
print(param)
x = np.linspace(10, 14, 100)      # Linear spacing of 100 elements between 10 and 14.
pdf_fitted = stats.norm.pdf(x, *param)    # Use the fitted paramters to create the y datapoints

print(pdf_fitted)

# Plot the histogram again
df_pow.hist(bins=100, figsize=(6,6))

# Plot some fancy text to show us what the paramters of the distribution are (mean and standard deviation)
plt.text(x=np.min(df_pow), y=50, s=r"$\mu=%0.1f$" % param[0] + "\n" + r"$\sigma=%0.1f$" % param[1], color='r')

# Plot a line of the fitted distribution over the top
plt.plot(x, pdf_fitted, color='r')

# Standard plot stuff
plt.xlabel("Feature value")
plt.title("Histogram with fitted normal distribution")
plt.show()

# I really want to find a way to plot normal distribution above my histogram but I can't. For some reason all these plots have
# range (0,1) on y axis or they are not presented as a line.
```

```python
# your code here
# add cells as needed
fig, ax = plt.subplots(1, 2, figsize=(15,8))

labels = df['SalePrice'].apply(lambda x: x/1000)
ax[0].scatter(data=df, x='SalePrice', y='Neighborhood')
ax[0].set_xticklabels(labels, ha='left')

labels = df['SalePriceAdjusted'].apply(lambda x: round(x, 2))
ax[1].scatter(data=df, x='SalePriceAdjusted', y='Neighborhood')
ax[1].set_xticklabels(labels, ha='left')

plt.subplots_adjust(wspace = 0.2)
plt.show()

'''
Neighborhood 'Crawfor' has some outliers on the upper end but most of the houses in this area have average price.
Neighborhood 'Sawyer' also has some outliers on the upper end but average prices of the most of the houses in this area are
higher than in 'Crawfor', also price range in this area is quite wide.
Most of the low-priced neighborhoods such as 'Edwards', 'Brikside', 'Mitchel' etc don't have such significant outliers as
neighborhoods mentioned earlier.
'''
```

```python
sns.pairplot(df[['GrLivArea', 'LotArea', 'GarageArea', 'SalePrice']])
plt.show()

l = ['GrLivArea', 'LotArea', 'GarageArea']
for n in l:
    print(n, df['SalePrice'].corr(df[n]))

'''
As per plots below lot area hardly influence the sales price but ground living area has almost linear regresion with sales price.
Also garage area has significant impact on sales price although it's not as linear as ground living area. These conclustions
are also supported by correlation coefficient.
'''
```

```python
df['TotalBath'] = df['FullBath'] + df['HalfBath'] + df['BsmtFullBath'] + df['BsmtHalfBath']
df['TotalRooms'] = df['BedroomAbvGr'] + df['KitchenAbvGr'] + df['TotalBath']
df.head()
```

```python
plt.scatter(data=df, y='SalePrice', x='TotalRooms')
plt.show()
```

```python
sns.pairplot(df[['BedroomAbvGr', 'KitchenAbvGr', 'TotalBath', 'SalePrice']])
plt.show()
l = ['BedroomAbvGr', 'KitchenAbvGr', 'TotalBath', 'TotalRooms']
for n in l:
    print(n, df['SalePrice'].corr(df[n]))
```

```python
'''
There is a tendency that total number of rooms has a positive impact on a house price but only up to certain number of bedrooms.
For example, houses with 11 rooms are on the lower end of price range in our data set. There are might be some other factors
that overpower benefits of total number of rooms, it also might be that there is certain maximum number of rooms which the most
common household prefers.
If we look on the pairplots above we can see that total number of bathrooms has the biggest correlation with sales price.
Number of kitchen is not really informative in terms of analysis because there is not such a big range of possibilities,
most of the houses will have one kitchen and most of the households will require just one kitchen.
These conclusions are confirmed by calculated correlation coefficients.
'''
```

```python
df['HouseAge'] = df['YearBuilt'].apply(lambda x: 2010 - x)
df['RepairAge'] = df['YearRemodAdd'].apply(lambda x: 2010 - x)
# Not 2020 but 2010 because data set is old, latest sales are from 2010

df['UltimateAge'] = df[['HouseAge', 'RepairAge']].min(axis=1)
```

```python
sns.pairplot(df[['HouseAge', 'RepairAge', 'UltimateAge', 'SalePrice']])
plt.show()
df['SalePrice'].corr(df['UltimateAge'])
```

```python
'''
As per pairplots above there is a negative correlation between age of the house or the age of reinovation and sales price.
It is also visible from calculated correlation coefficient.
'''
```

## Bonus Challenge 2 - Exploring Data with Correlation Heatmap

Now you have explored data visualizations with certain fields based on common sense. In the dataset there are many other fields that you are not sure whether they are important factors for the sale price. What is the best way to explore those fields without investigating them individually?

Making scatter matrix is not an option here because there are too many fields which makes it extremely time consuming to create scatter matrix. One option you have is to create a heatmap. Heatmaps are much less expensive to create than scatter matrixes. You can use heatmaps to visualize the pairwise correlations between each two variables.

Here is a [reference](https://seaborn.pydata.org/examples/many_pairwise_correlations.html) you can use to learn how to creat the pairwise correlation heatmap. Your heatmap should look like below:

![Corr Heatmap](heatmap.png)

```python
# your code here

# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": 0.8})
```

```python
# Based on heatmap we should also check out impact of overal material and finish quality on sales price.
plt.scatter(data=df, y='SalePrice', x='OverallQual')

# There is a strong positive correlation between quality of materials and house price.
```

```python
# Repetition of prior calculations for easy access while writing final conclusion.

l = ['GrLivArea', 'LotArea', 'GarageArea', 'BedroomAbvGr', 'KitchenAbvGr', 'TotalBath', 'UltimateAge', 'OverallQual']
for n in l:
    print(n, df['SalePrice'].corr(df[n]))
```

In your heatmap, you can easily identify the highly correlated (either positively or negatively) variables by looking for the grids with darker colors.

Now based on your findings from the explorations, summarize and present your stories.

Use the following format to present each factor:

```python
# your responses here
# add cells as needed

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
df_plot = df[['OverallQual', 'SalePrice']].groupby(df['OverallQual']).mean()

# Plot 1
df_plot.plot(x='OverallQual', ax=ax[0])
ax[0].set_title('Overall Quality of Materials & Finish vs Average Price')

# Plot 2
data = df['OverallQual'].groupby(df['OverallQual']).count().nlargest(5)
ax[1].pie(data, labels=data.index, autopct='%1.1f%%')
ax[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax[1].set_title('Overall Quality of Materials & Finish, market share (top 5)')
plt.show()
```

Overall quality of materials and finish has direct impact on house prices but it's important to keep in mind that the biggest market share is presented by houses with 5-7 bathrooms, so depending on a strategy it might be more profitable to invest in houses with lower prices per unit but wider presented on a market.

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1
ax[0].scatter(data=df, x='GrLivArea', y='SalePrice')
ax[0].set_title('Ground living area vs Average Price')

# Plot 2
ax[1].hist(df['GrLivArea'], weights=np.ones(len(df['GrLivArea'])) / len(df['GrLivArea']))
ax[1].set_title('Ground living area, market share')
plt.show()

# Here and for all numeric fields below: it would be better to split field in bins and then use this for pie chart but I didn't
# have time to look into this solution, will try to finish during this week.
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
df_plot = df[['TotalBath', 'SalePrice']].groupby(df['TotalBath']).mean()

# Plot 1
df_plot.plot(x='TotalBath', ax=ax[0])
ax[0].set_title('Total number of bathrooms vs Average Price')

# Plot 2
data = df['TotalBath'].groupby(df['TotalBath']).count()
ax[1].pie(data, labels=data.index, autopct='%1.1f%%')
ax[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax[1].set_title('Total number of bathrooms, market share')

plt.subplots_adjust(wspace = 0.2)
plt.show()
```

In case of total number of bathrooms impact on price is again not exactly direct (i.e. linear).
Top priced houses are houses with 5 bathrooms but their market share is less that 1%. It might be better to consider average houses for investments.

```python
cond_1 = df['TotalBath'].isin([2, 3])
cond_2 = df['GarageCars'].isin([1, 2])
cond_3 = df['OverallQual'].isin([5, 6, 7])
cond_4 = df['GrLivArea'].between(1000, 2000)


selected = df.loc[cond_1 & cond_2 & cond_3 & cond_4]
income = selected['SalePrice'].sum()

selected['SalePrice'].sort_values().reset_index()['SalePrice'].plot()
print(f'{round(len(selected)/len(df)*100)}%')
print(selected['SalePrice'].min(), selected['SalePrice'].max())
print(df['SalePrice'].max())
print(income)
plt.show()
```

```python
q75 = df['SalePrice'].quantile(0.75)
top25 = df[df['SalePrice'] >= q75]
income = top25['SalePrice'].sum()
print(income)
```

If we select houses, combining conditions reviewed above, we get 42% of market with price range from USD66.5k to USD335k. Total income (not profit) would be USD102m.
If we select top 25% of houses based on sales price then our income would be USD105.6m.
Based on this analysis it might be more profitable to go with less market share but with higher prices, although we don't have information regarding any related expenses.
